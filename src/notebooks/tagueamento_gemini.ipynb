# Criação da função de tagueamento

```python
import os

from dotenv import load_dotenv

import google.generativeai as genai
import json
```

```python
load_dotenv()
GEMINI_API_KEY = os.getenv('GEMINI_API_KEY')
genai.configure(api_key=GEMINI_API_KEY)
model = genai.GenerativeModel('gemini-pro')
```

Na função abaixo, utilizamos o Gemini para ler o documento e gerar as tags. Isso é feito através do prompt que ordena ele a criar tags, baseado no documento, e avaliar se o documento foi revogado ou não. A saida da função é um json com essas informações.



```python
def generate_content_gemini(document):
    initial_prompt = """
    Você é um assistente que ajuda a criar tags para documentos. Toda vez que eu enviar um documento, quero que você apenas gere uma única lista de palavras em python (ex: ['tag1', 'tag2', 'tag3']) relevantes com base no conteúdo do documento. Essa lista deve conter no mínimo 5 palavras/categorias gerais e mais algumas que sejam específicas do assunto do documento. Exemplo: 'Hoje irei comer uma pizza em um resturante', palavras na formatação esperada: ['comida', 'restaurantes', 'gastronomia', 'jantar', 'almoço', 'pizza', 'restaurante']. O output gerado deve ser um objeto, dessa forma:
    {
      "tags": ['tag1', 'tag2', 'tag3'],
      "revogado": true/false
    }. Caso você identifique que o documento foi revogado coloque "true" em "revogado", caso contrário coloque "false".
    """
    prompt = initial_prompt + "\n\nDocumento: " + document + "\n\nTags:"

    if not document or document.strip() == '':
      raise ValueError("O documento fornecido está vazio ou é inválido.")

    response = model.generate_content(prompt)

    try:
        result = json.loads(response.text)
    except json.JSONDecodeError:
        raise ValueError("O formato do response não está em JSON válido.")

    return result
```
